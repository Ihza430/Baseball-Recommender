# Data Cleaning

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
bat = pd.read_csv('../data/lahman_database/Batting.csv')
pitch = pd.read_csv('../data/lahman_database/Pitching.csv')
```

```python
bat.head()
```

```python
pitch.head()
```

## Bat Cleaning

```python
bat = bat[bat['yearID']>= 2010]
bat.head()
```

```python
bat.drop(columns = ['stint', 'G', '2B', '3B', 'SB', 'CS','IBB', 'HBP', 'GIDP', 'SH','SF'], inplace = True)
bat.head()
```

```python
bat.isnull().sum()
```

```python
bat.info()
```

```python
bat['BA'] = bat['H'] / bat['AB']
```

```python
bat.head()
```

```python
bat.fillna(0, inplace = True)
```

```python
bat['BA'] = np.round(bat['BA'], 3)
```

```python
bat.head()
```

```python
bat.shape
```

## Save Clean Bat Data

```python
bat.to_csv('../data/clean/bat.csv')
```

## Pitch Cleaning
